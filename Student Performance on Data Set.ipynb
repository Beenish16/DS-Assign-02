```python
#VISUALIZATION

#this shows the distribution of student's age, studytime and traveltime

plt.figure(figsize = (14,6))

plt.subplot(1, 3, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.
df['age'].plot(kind='hist')
plt.xlabel('age')
plt.title('age')
plt.tight_layout()

plt.subplot(1, 3, 2)
df['traveltime'].plot(kind='hist')
plt.xlabel('traveltime')
plt.title('traveltime')
plt.tight_layout()

plt.subplot(1, 3, 3) 
df['studytime'].plot(kind='hist')
plt.xlabel('studytime')
plt.title('studytime')
plt.tight_layout()
```

```python
#this shows the statistical analysis of student's condition in health, absences and G1

plt.figure(figsize = (14,6))

plt.subplot(1, 3, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.
sns.boxplot(data= df, x= 'health')
plt.title('health')
plt.tight_layout()

plt.subplot(1, 3, 2) 
sns.boxplot(data= df, x= 'absences')
plt.title('absences')
plt.tight_layout()

plt.subplot(1, 3, 3) 
sns.boxplot(data= df, x= 'G1')
plt.title('G1')
plt.tight_layout()
```

```python
#TO CHECK THE GENDER PERFORMANCE ON FAMILY RELATIONS , MOTHER EDUCATION AND FATHER EDUCATION

famrel_filtered = df.query('famrel > 3')
Medu_filtered = df.query('Medu > 3')
Fedu_filtered = df.query('Fedu > 3')

famrel_stat = famrel_filtered.groupby('sex')['famrel'].count()
Medu_stat = Medu_filtered.groupby('sex')['Medu'].count()
Fedu_stat = Fedu_filtered.groupby('sex')['Fedu'].count()
```

```python
plt.figure(figsize = (14,6))

plt.subplot(1, 3, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.
famrel_stat.plot(kind= 'pie', autopct= "%.0f%%")
plt.title('famrel')
plt.ylabel('')
plt.tight_layout()

plt.subplot(1, 3, 2)
Medu_stat.plot(kind= 'pie', autopct= "%.0f%%")
plt.title('Medu')
plt.ylabel('')
plt.tight_layout()

plt.subplot(1, 3, 3)
Fedu_stat.plot(kind= 'pie', autopct= "%.0f%%")
plt.title('Fedu')
plt.ylabel('')
plt.tight_layout()
```
